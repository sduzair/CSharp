```polyglot-notebook
public static class FizzBuzz
{
   private static readonly Dictionary<int, string> _predicates = new() {
        { 3, "Fizz" },
        { 5, "Buzz" },
        { 7, "Bazz" }
   };

   public static string GetWord(int number)
   {
        string word = string.Empty;

        foreach (int key in _predicates.Keys)
        {
            if (number % key == 0)
            {
                word += _predicates[key];
            }
        }

        return word;
   } 
}
```

```polyglot-notebook
for (int i = 0; i < 107; i++)
{
    string word = FizzBuzz.GetWord(i);
    if (string.IsNullOrEmpty(word)) Console.WriteLine(i);
    else Console.WriteLine(word);
}
```
